## 0. 들여쓰기 ― 코드 묶음의 표시
파이썬은 **들여쓰기(앞 공백)** 로 "어디부터 어디까지가 한 묶음인지"를 구분합니다. `if`·`for`·`def` 처럼 묶음을 여는 줄은 끝에 콜론(`:`)을 붙이고, 그 아래 묶음에 속하는 줄들은 **같은 칸수만큼 들여씁니다**(보통 공백 4칸). Colab에서는 콜론 다음 줄에서 자동으로 들여써집니다.

들여쓰기가 어긋나면 오류가 나므로, **한 묶음 안에서는 들여쓰기 칸수를 똑같이** 맞추는 것이 중요합니다.

```python
# 콜론(:) 아래 들여쓴 두 줄이 if 에 속한 한 묶음입니다.
temperature = 30
if temperature >= 28:
    print("덥습니다.")        # if 에 속함 (들여쓰기됨)
    print("에어컨을 켭니다.")   # if 에 속함 (같은 칸수)
print("측정 완료.")           # 들여쓰기 없음 → if 와 무관, 항상 실행
# 출력:
# 덥습니다.
# 에어컨을 켭니다.
# 측정 완료.
```

## 1. 조건문 ― `if`
`if` 는 "조건이 참(`True`)일 때만 실행"하는 갈림길입니다. 조건은 비교 연산의 결과(`True`/`False`)입니다.

```python
score = 85

if score >= 60:
    print("합격입니다.")   # 조건이 참이므로 출력됨
# 출력:
# 합격입니다.
```

### 1-1) `if` / `else`
`else` 는 "조건이 거짓일 때" 실행할 묶음입니다. 둘 중 하나는 반드시 실행됩니다.

```python
score = 45

if score >= 60:
    print("합격입니다.")
else:
    print("불합격입니다.")   # 45 는 60 미만이므로 이쪽이 실행됨
# 출력:
# 불합격입니다.
```

### 1-2) `if` / `elif` / `else` ― 여러 갈래
갈림길이 셋 이상이면 `elif`(else if)를 사이에 넣습니다. **위에서부터 차례로 검사**해, 처음으로 참이 되는 묶음 하나만 실행하고 나머지는 건너뜁니다. 아래는 점수로 등급을 매기는 예입니다.

```python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:      # 90 미만이면서 80 이상
    grade = "B"
elif score >= 70:
    grade = "C"
else:                  # 위 조건이 모두 거짓일 때
    grade = "F"

print(f"점수 {score} → 등급 {grade}")   # 점수 85 → 등급 B
```

```python
# 연습문제 1
age = 22

if age >= 20:
    print("성인")
elif age >= 14:
    print("청소년")
else:
    print("어린이")
# 출력: 성인
```

## 2. 조건 만들기 ― 비교·논리 연산 복습
조건 자리에는 비교(`>`, `>=`, `==`, `!=` 등)와 논리(`and`, `or`, `not`)를 함께 쓸 수 있습니다. `and` 는 둘 다 참, `or` 는 하나라도 참, `not` 은 반대입니다.

```python
age = 22
has_ticket = True

# and: 두 조건이 모두 참이어야 입장
if age >= 19 and has_ticket:
    print("입장 가능")     # 둘 다 참 → 출력
else:
    print("입장 불가")
# 출력: 입장 가능

# or: 하나라도 참이면 할인
is_student = True
is_member = False
if is_student or is_member:
    print("할인 대상")     # 학생이 참 → 출력
# 출력: 할인 대상
```

### 2-1) 범위 조건과 멤버십
`a <= x <= b` 처럼 **범위**를 한 번에 적을 수 있고, `in` 으로 어떤 값이 리스트 안에 있는지 검사할 수 있습니다.

```python
score = 75
if 70 <= score < 80:
    print("70점대입니다.")   # 75 는 70 이상 80 미만 → 출력
# 출력: 70점대입니다.

fruit = "사과"
basket = ["사과", "바나나", "포도"]
if fruit in basket:
    print("바구니에 있습니다.")   # "사과" 가 리스트에 있음 → 출력
# 출력: 바구니에 있습니다.
```

### 2-2) 중첩 조건 ― 조건 안의 조건
`if` 묶음 안에 또 `if` 를 넣어 단계적으로 판단할 수 있습니다. 안쪽은 한 단계 **더 들여씁니다**(공백 8칸). 아래는 로그인 검증 예입니다.

```python
user_id = "hknu"
password = "1234"

if user_id == "hknu":          # 1단계: 아이디 확인
    if password == "1234":     # 2단계: 비밀번호 확인 (더 들여씀)
        print("로그인 성공")
    else:
        print("비밀번호가 틀렸습니다.")
else:
    print("없는 아이디입니다.")
# 출력: 로그인 성공
```

## 3. 반복문 ― `for` 와 `range`
`for` 는 여러 항목을 **하나씩 차례로 꺼내며** 같은 작업을 되풀이합니다. `range(n)` 은 0부터 n-1까지의 숫자를 만들어 주므로, "n번 반복"에 자주 씁니다.

```python
# range(5) → 0, 1, 2, 3, 4 (5는 포함하지 않음)
for i in range(5):
    print(i)
# 출력:
# 0
# 1
# 2
# 3
# 4
```

```python
# range(시작, 끝): 시작부터 끝-1까지
for i in range(1, 6):
    print(f"{i}번째 줄")
# 출력:
# 1번째 줄
# 2번째 줄
# 3번째 줄
# 4번째 줄
# 5번째 줄

# range(시작, 끝, 간격)
for i in range(0, 11, 2):   # 0부터 10까지 2씩
    print(i, end=" ")        # end=" " : 줄바꿈 대신 공백
# 출력: 0 2 4 6 8 10 
```

### 3-1) 구구단 ― `for` 활용
`for` 로 반복하면 비슷한 줄을 여러 번 직접 쓰지 않아도 됩니다. 아래는 3단을 출력합니다.

```python
dan = 3
for i in range(1, 10):     # 1부터 9까지
    print(f"{dan} x {i} = {dan * i}")
# 출력:
# 3 x 1 = 3
# 3 x 2 = 6
# ... (중략) ...
# 3 x 9 = 27
```

```python
# 연습문제 3
for n in range(1, 11):
    print(f"{n} -> {n * n}")
# 출력:
# 1 -> 1
# 2 -> 4
# ... (중략) ...
# 10 -> 100
```

## 4. 리스트·딕셔너리 순회
`for` 는 `range` 뿐 아니라 **리스트·문자열·딕셔너리**의 항목도 하나씩 꺼낼 수 있습니다.

```python
fruits = ["사과", "바나나", "체리"]
for fruit in fruits:        # 리스트 항목을 하나씩 fruit 에 담음
    print(f"과일: {fruit}")
# 출력:
# 과일: 사과
# 과일: 바나나
# 과일: 체리
```

### 4-1) 번호가 필요하면 `enumerate`
순서 번호도 함께 쓰고 싶을 땐 `enumerate` 를 씁니다. (번호, 값) 두 개를 한 번에 꺼냅니다.

```python
fruits = ["사과", "바나나", "체리"]
for idx, fruit in enumerate(fruits, start=1):   # 1번부터 매김
    print(f"{idx}번: {fruit}")
# 출력:
# 1번: 사과
# 2번: 바나나
# 3번: 체리
```

### 4-2) 딕셔너리 순회
딕셔너리는 `.items()` 로 **키와 값**을 한 번에 꺼낼 수 있습니다.

```python
student = {"name": "지민", "age": 22, "major": "AI"}
for key, value in student.items():
    print(f"{key} : {value}")
# 출력:
# name : 지민
# age : 22
# major : AI
```

```python
# 연습문제 4
scores = [88, 52, 73, 95, 40]
for s in scores:
    if s >= 60:
        print(f"{s}점 -> 통과")
    else:
        print(f"{s}점 -> 재시험")
# 출력:
# 88점 -> 통과
# 52점 -> 재시험
# 73점 -> 통과
# 95점 -> 통과
# 40점 -> 재시험
```

## 5. `while` ― 조건이 참인 동안 반복
`for` 가 "정해진 항목만큼" 반복한다면, `while` 은 **조건이 참인 동안 계속** 반복합니다. 반복을 멈추려면 조건이 언젠가 거짓이 되도록 **값을 바꿔 주는 줄**이 꼭 있어야 합니다(없으면 무한 반복).

```python
count = 3
while count > 0:           # count 가 0보다 큰 동안 반복
    print(f"발사까지 {count}")
    count = count - 1       # 매번 1씩 줄임 (이 줄이 없으면 무한 반복!)
print("발사!")
# 출력:
# 발사까지 3
# 발사까지 2
# 발사까지 1
# 발사!
```

### 5-1) `break` ― 반복 중단, `continue` ― 다음으로 건너뛰기
`break` 는 반복을 즉시 멈추고, `continue` 는 이번 회차만 건너뛰고 다음 회차로 넘어갑니다.

```python
# break: 5를 만나면 즉시 멈춤
for n in range(1, 10):
    if n == 5:
        break
    print(n, end=" ")
# 출력: 1 2 3 4 
```

```python
# continue: 홀수는 건너뛰고 짝수만 출력
for n in range(1, 11):
    if n % 2 == 1:      # 홀수이면
        continue         # 아래를 건너뛰고 다음 n 으로
    print(n, end=" ")
# 출력: 2 4 6 8 10 
```

```python
# 연습문제 5
num = 10
while num > 0:
    print(num)
    num = num - 2
# 출력:
# 10
# 8
# 6
# 4
# 2
```

## 6. 자주 쓰는 패턴 ― 누적(합계)과 카운트
반복문에서 가장 흔한 일은 **값을 모아 합치기**와 **개수 세기**입니다. 반복 밖에서 변수를 0으로 시작해 두고, 반복 안에서 더하거나 1씩 늘립니다.

`total = total + x` 는 `total += x` 로 짧게 쓸 수 있습니다(`+=`, `-=` 등).

```python
scores = [88, 92, 79, 95, 60]

total = 0                 # 합계를 담을 그릇 (0으로 시작)
for s in scores:
    total += s             # total = total + s 와 같음
print("합계:", total)      # 합계: 414
print("평균:", total / len(scores))   # 평균: 82.8
```

```python
scores = [88, 52, 73, 95, 40]

pass_count = 0            # 통과한 학생 수
for s in scores:
    if s >= 60:
        pass_count += 1    # 조건에 맞을 때만 1 증가
print(f"통과: {pass_count}명")   # 통과: 3명
```

### 연습문제 6
아래 `nums` 에서 **짝수만 골라 합**을 구해 출력하세요.
- `total` 을 0으로 시작하고, `for` 로 하나씩 검사합니다.
- 짝수인지(`% 2 == 0`)를 `if` 로 확인해 맞을 때만 더합니다.

```python
nums = [3, 8, 5, 12, 7, 6, 10]
total = 0
for n in nums:
    if n % 2 == 0:
        total += n
print("짝수의 합:", total)   # 짝수의 합: 36  (8+12+6+10)
```

## 7. (가볍게) 리스트 컴프리헨션
"리스트를 만드는 `for`" 를 한 줄로 줄여 쓰는 문법입니다. 아직 익숙하지 않으면 위의 `for` 방식으로 충분합니다. 코드에서 자주 보이므로 **눈에 익혀 두는** 정도면 됩니다.

```python
# 일반 for 로 제곱 리스트 만들기
squares = []
for n in range(1, 6):
    squares.append(n * n)
print(squares)   # [1, 4, 9, 16, 25]

# 같은 결과를 컴프리헨션 한 줄로
squares2 = [n * n for n in range(1, 6)]
print(squares2)  # [1, 4, 9, 16, 25]

# 조건을 붙일 수도 있습니다 (짝수만)
evens = [n for n in range(1, 11) if n % 2 == 0]
print(evens)     # [2, 4, 6, 8, 10]
```

## 8. 함수 ― 작업을 묶어 이름 붙이기
함수는 여러 줄의 작업을 **하나로 묶어 이름을 붙인 것**입니다. 한 번 정의해 두면 이름으로 **여러 번 다시 호출**할 수 있어, 같은 코드를 반복해 쓰지 않아도 됩니다.

- `def 이름(매개변수):` 로 정의하고, 묶음은 들여씁니다.
- 호출은 `이름(값)` 으로 합니다.
- `return` 으로 결과 값을 돌려줍니다(없으면 `None`).

```python
# 가장 단순한 함수: 인사말 출력
def greet():
    print("안녕하세요, 환영합니다!")

greet()        # 호출 → 안녕하세요, 환영합니다!
greet()        # 다시 호출 → 안녕하세요, 환영합니다!
# 출력:
# 안녕하세요, 환영합니다!
# 안녕하세요, 환영합니다!
```

### 8-1) 매개변수와 반환값
괄호 안의 **매개변수**로 값을 받아, `return` 으로 결과를 돌려줍니다. 돌려받은 값은 변수에 저장해 쓸 수 있습니다.

```python
def add(a, b):
    result = a + b
    return result          # 계산 결과를 돌려줌

x = add(3, 5)              # 돌려받은 값을 x 에 저장
print(x)                   # 8
print(add(10, 20))         # 30  (바로 출력도 가능)
```

### 8-2) 기본값 매개변수
매개변수에 `=값` 으로 **기본값**을 정하면, 호출할 때 그 값을 생략할 수 있습니다. 아래 환율 함수는 통화를 생략하면 원화로 계산합니다.

```python
def to_won(amount, rate=1300):    # rate 의 기본값은 1300
    return amount * rate

print(to_won(10))            # 13000   (기본 환율 1300 적용)
print(to_won(10, 1450))      # 14500   (환율을 직접 지정)
```

### 8-3) 여러 값 반환
`return` 에 값을 쉼표로 나열하면 **여러 값을 한 번에** 돌려줄 수 있고, 받는 쪽도 변수 여러 개로 나눠 받습니다.

```python
def min_max(numbers):
    return min(numbers), max(numbers)   # 최솟값, 최댓값을 함께 반환

low, high = min_max([88, 92, 79, 95, 60])
print("최저:", low)    # 최저: 60
print("최고:", high)   # 최고: 95
```

### 8-4) docstring ― 함수 설명문
함수 첫 줄에 따옴표 세 개(`"""..."""`)로 설명을 적어 두면, 다른 사람(과 미래의 나)이 함수의 용도를 알 수 있습니다. 이를 docstring 이라 합니다.

```python
def average(numbers):
    """숫자 리스트를 받아 평균을 돌려줍니다."""
    return sum(numbers) / len(numbers)

print(average([80, 90, 100]))   # 90.0
print(average.__doc__)          # 숫자 리스트를 받아 평균을 돌려줍니다.
```

   - 90 이상 `"A"`, 80 이상 `"B"`, 70 이상 `"C"`, 그 외 `"F"`
2. 아래 `students` 의 각 학생에 대해 `grade_of` 로 등급을 구해
   `"지민: 88점 -> B"` 형식으로 한 줄씩 출력하세요.
   (힌트: `for` 로 딕셔너리를 순회하고, 함수를 호출해 등급을 얻습니다.)

```python
students = [
    {"name": "지민", "score": 88},
    {"name": "현우", "score": 95},
    {"name": "수아", "score": 72},
    {"name": "민재", "score": 54},
]

# 1) 등급 함수 정의
def grade_of(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "F"

# 2) 반복하며 등급 출력
for s in students:
    g = grade_of(s["score"])
    print(f"{s['name']}: {s['score']}점 -> {g}")
# 출력:
# 지민: 88점 -> B
# 현우: 95점 -> A
# 수아: 72점 -> C
# 민재: 54점 -> F
```

### 연습문제 9 ― 통계 함수 (도전)
숫자 리스트를 받아 **개수·합계·평균**을 한 번에 돌려주는 함수 `summary` 를 만드세요.
- `return count, total, total / count` 처럼 세 값을 함께 돌려줍니다.
- 아래 데이터로 호출해 세 변수로 나눠 받은 뒤 보기 좋게 출력하세요.

```python
data = [12, 7, 25, 33, 18, 9]

# 1) summary 함수 정의
def summary(numbers):
    """숫자 리스트의 개수·합계·평균을 돌려줍니다."""
    count = len(numbers)
    total = sum(numbers)
    return count, total, total / count

# 2) 호출하고 결과를 세 변수로 받아 출력
cnt, tot, avg = summary(data)
print(f"개수: {cnt}")            # 개수: 6
print(f"합계: {tot}")            # 합계: 104
print(f"평균: {avg:.2f}")        # 평균: 17.33
```

## 정리
- **`if` / `elif` / `else`** : 조건에 따라 다른 묶음을 실행합니다. 들여쓰기로 묶음을 구분합니다.
- **`for` + `range`** : 항목을 차례로 꺼내며 반복합니다. 리스트·딕셔너리도 순회할 수 있습니다.
- **`while`** : 조건이 참인 동안 반복하며, `break`(중단)·`continue`(건너뛰기)로 흐름을 제어합니다.
- **누적·카운트**(`total += x`, `count += 1`)는 반복문의 대표 패턴입니다.
- **함수**(`def`)는 작업을 묶어 이름 붙인 것으로, 매개변수로 값을 받고 `return` 으로 결과를 돌려줍니다.
