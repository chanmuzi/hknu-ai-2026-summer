## 0. 들여쓰기 ― 코드 묶음의 표시
파이썬은 **들여쓰기(앞 공백)** 로 "어디부터 어디까지가 한 묶음인지"를 구분합니다. `if`·`for`·`def` 처럼 묶음을 여는 줄은 끝에 콜론(`:`)을 붙이고, 그 아래 묶음에 속하는 줄들은 **같은 칸수만큼 들여씁니다**(보통 공백 4칸). Colab에서는 콜론 다음 줄에서 자동으로 들여써집니다.

들여쓰기가 어긋나면 오류가 나므로, **한 묶음 안에서는 들여쓰기 칸수를 똑같이** 맞추는 것이 중요합니다.

```python
# 콜론(:) 아래 들여쓴 두 줄이 if 에 속한 한 묶음
temperature = 30
if temperature >= 28:
    print("덥습니다.")
    print("에어컨을 켭니다.")
print("측정 완료.")           # 들여쓰기 없음 → if 와 무관, 항상 실행
```

## 1. 조건문 ― `if`
`if` 는 "조건이 참(`True`)일 때만 실행"하는 갈림길입니다. 조건은 비교 연산의 결과(`True`/`False`)입니다.

```python
score = 85
if score >= 60:
    print("합격입니다.")   # 조건이 참일 때만 실행
```

### 1-1) `if` / `else`
`else` 는 "조건이 거짓일 때" 실행할 묶음입니다. 둘 중 하나는 반드시 실행됩니다.

```python
score = 45
if score >= 60:
    print("합격입니다.")
else:
    print("불합격입니다.")   # 조건이 거짓이면 이쪽
```

### 1-2) `if` / `elif` / `else` ― 여러 갈래
갈림길이 셋 이상이면 `elif`(else if)를 사이에 넣습니다. **위에서부터 차례로 검사**해, 처음으로 참이 되는 묶음 하나만 실행하고 나머지는 건너뜁니다. 아래는 점수로 등급을 매기는 예입니다.

```python
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:      # 90 미만이면서 80 이상
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"
print(f"점수 {score} → 등급 {grade}")
```

## 2. 조건 만들기 ― 비교·논리 연산 복습
조건 자리에는 비교(`>`, `>=`, `==`, `!=` 등)와 논리(`and`, `or`, `not`)를 함께 쓸 수 있습니다. `and` 는 둘 다 참, `or` 는 하나라도 참, `not` 은 반대입니다.

```python
age = 22
has_ticket = True
if age >= 19 and has_ticket:   # and: 둘 다 참
    print("입장 가능")
else:
    print("입장 불가")

is_student = True
is_member = False
if is_student or is_member:     # or: 하나라도 참
    print("할인 대상")
```

### 2-1) 범위 조건과 멤버십
`a <= x <= b` 처럼 **범위**를 한 번에 적을 수 있고, `in` 으로 어떤 값이 리스트 안에 있는지 검사할 수 있습니다.

```python
score = 75
if 70 <= score < 80:            # 범위 조건
    print("70점대입니다.")

fruit = "사과"
basket = ["사과", "바나나", "포도"]
if fruit in basket:            # in: 리스트에 있는지
    print("바구니에 있습니다.")
```

```python
user_id = "hknu"
password = "1234"
if user_id == "hknu":          # 1단계: 아이디
    if password == "1234":     # 2단계: 비밀번호 (더 들여씀)
        print("로그인 성공")
    else:
        print("비밀번호가 틀렸습니다.")
else:
    print("없는 아이디입니다.")
```

## 3. 반복문 ― `for` 와 `range`
`for` 는 여러 항목을 **하나씩 차례로 꺼내며** 같은 작업을 되풀이합니다. `range(n)` 은 0부터 n-1까지의 숫자를 만들어 주므로, "n번 반복"에 자주 씁니다.

```python
# range(5) → 0,1,2,3,4 (5는 미포함)
for i in range(5):
    print(i)
```

```python
for i in range(1, 6):           # 시작~끝-1
    print(f"{i}번째 줄")

for i in range(0, 11, 2):       # 0부터 10까지 2씩
    print(i, end=" ")            # end=" " : 줄바꿈 대신 공백
```

## 4. 리스트·딕셔너리 순회
`for` 는 `range` 뿐 아니라 **리스트·문자열·딕셔너리**의 항목도 하나씩 꺼낼 수 있습니다.

```python
fruits = ["사과", "바나나", "체리"]
for fruit in fruits:            # 리스트 항목을 하나씩
    print(f"과일: {fruit}")
```

### 4-1) 번호가 필요하면 `enumerate`
순서 번호도 함께 쓰고 싶을 땐 `enumerate` 를 씁니다. (번호, 값) 두 개를 한 번에 꺼냅니다.

```python
fruits = ["사과", "바나나", "체리"]
for idx, fruit in enumerate(fruits, start=1):   # 번호와 값을 함께
    print(f"{idx}번: {fruit}")
```

### 4-2) 딕셔너리 순회
딕셔너리는 `.items()` 로 **키와 값**을 한 번에 꺼낼 수 있습니다.

```python
student = {"name": "지민", "age": 22, "major": "AI"}
for key, value in student.items():   # 키와 값을 함께
    print(f"{key} : {value}")
```

## 5. `while` ― 조건이 참인 동안 반복
`for` 가 "정해진 항목만큼" 반복한다면, `while` 은 **조건이 참인 동안 계속** 반복합니다. 반복을 멈추려면 조건이 언젠가 거짓이 되도록 **값을 바꿔 주는 줄**이 꼭 있어야 합니다(없으면 무한 반복).

```python
count = 3
while count > 0:                # 조건이 참인 동안 반복
    print(f"발사까지 {count}")
    count = count - 1            # 이 줄이 없으면 무한 반복!
print("발사!")
```

### 5-1) `break` ― 반복 중단, `continue` ― 다음으로 건너뛰기
`break` 는 반복을 즉시 멈추고, `continue` 는 이번 회차만 건너뛰고 다음 회차로 넘어갑니다.

```python
# break: 5를 만나면 즉시 멈춤
for n in range(1, 10):
    if n == 5:
        break
    print(n, end=" ")
# 출력: 1 2 3 4 
```

```python
# continue: 홀수는 건너뛰고 짝수만 출력
for n in range(1, 11):
    if n % 2 == 1:
        continue
    print(n, end=" ")
```

## 6. 자주 쓰는 패턴 ― 누적(합계)과 카운트
반복문에서 가장 흔한 일은 **값을 모아 합치기**와 **개수 세기**입니다. 반복 밖에서 변수를 0으로 시작해 두고, 반복 안에서 더하거나 1씩 늘립니다.

`total = total + x` 는 `total += x` 로 짧게 쓸 수 있습니다(`+=`, `-=` 등).

```python
scores = [88, 92, 79, 95, 60]
total = 0                       # 합계 그릇 (0으로 시작)
for s in scores:
    total += s                   # total = total + s
print("합계:", total)
print("평균:", total / len(scores))
```

```python
scores = [88, 52, 73, 95, 40]
pass_count = 0                  # 통과 학생 수
for s in scores:
    if s >= 60:
        pass_count += 1
print(f"통과: {pass_count}명")
```

## 7. (가볍게) 리스트 컴프리헨션
"리스트를 만드는 `for`" 를 한 줄로 줄여 쓰는 문법입니다. 아직 익숙하지 않으면 위의 `for` 방식으로 충분합니다. 코드에서 자주 보이므로 **눈에 익혀 두는** 정도면 됩니다.

```python
# 일반 for 로 제곱 리스트
squares = []
for n in range(1, 6):
    squares.append(n * n)
print(squares)

# 같은 결과를 컴프리헨션 한 줄로
squares2 = [n * n for n in range(1, 6)]
print(squares2)

# 조건을 붙일 수도 있음 (짝수만)
evens = [n for n in range(1, 11) if n % 2 == 0]
print(evens)
```

## 8. 함수 ― 작업을 묶어 이름 붙이기
함수는 여러 줄의 작업을 **하나로 묶어 이름을 붙인 것**입니다. 한 번 정의해 두면 이름으로 **여러 번 다시 호출**할 수 있어, 같은 코드를 반복해 쓰지 않아도 됩니다.

- `def 이름(매개변수):` 로 정의하고, 묶음은 들여씁니다.
- 호출은 `이름(값)` 으로 합니다.
- `return` 으로 결과 값을 돌려줍니다(없으면 `None`).

```python
# 가장 단순한 함수: 인사말 출력
def greet():
    print("안녕하세요, 환영합니다!")

greet()        # 호출
greet()        # 다시 호출
```

### 8-1) 매개변수와 반환값
괄호 안의 **매개변수**로 값을 받아, `return` 으로 결과를 돌려줍니다. 돌려받은 값은 변수에 저장해 쓸 수 있습니다.

```python
def add(a, b):
    return a + b               # 결과를 돌려줌

x = add(3, 5)                 # 돌려받은 값을 저장
print(x)                      # 8
print(add(10, 20))            # 바로 출력도 가능
```

### 8-2) 기본값 매개변수
매개변수에 `=값` 으로 **기본값**을 정하면, 호출할 때 그 값을 생략할 수 있습니다. 아래 환율 함수는 통화를 생략하면 원화로 계산합니다.

```python
def to_won(amount, rate=1300):    # rate 기본값 1300
    return amount * rate

print(to_won(10))            # 기본 환율 → 13000
print(to_won(10, 1450))      # 환율 지정 → 14500
```

### 8-3) 여러 값 반환
`return` 에 값을 쉼표로 나열하면 **여러 값을 한 번에** 돌려줄 수 있고, 받는 쪽도 변수 여러 개로 나눠 받습니다.

```python
def min_max(numbers):
    return min(numbers), max(numbers)   # 두 값을 함께 반환

low, high = min_max([88, 92, 79, 95, 60])
print("최저:", low)
print("최고:", high)
```

### 8-4) docstring ― 함수 설명문
함수 첫 줄에 따옴표 세 개(`"""..."""`)로 설명을 적어 두면, 다른 사람(과 미래의 나)이 함수의 용도를 알 수 있습니다. 이를 docstring 이라 합니다.

```python
def average(numbers):
    """숫자 리스트를 받아 평균을 돌려줍니다."""
    return sum(numbers) / len(numbers)

print(average([80, 90, 100]))   # 90.0
print(average.__doc__)          # 숫자 리스트를 받아 평균을 돌려줍니다.
```

## 정리
- **`if` / `elif` / `else`** : 조건에 따라 다른 묶음을 실행합니다. 들여쓰기로 묶음을 구분합니다.
- **`for` + `range`** : 항목을 차례로 꺼내며 반복합니다. 리스트·딕셔너리도 순회할 수 있습니다.

- **누적·카운트**(`total += x`, `count += 1`)는 반복문의 대표 패턴입니다.
- **함수**(`def`)는 작업을 묶어 이름 붙인 것으로, 매개변수로 값을 받고 `return` 으로 결과를 돌려줍니다.
