## 1. 표 데이터 만들기 — 리스트 안에 딕셔너리

엑셀 시트를 떠올려 봅시다. **한 줄(행)** 은 영화 한 편이고, **한 칸의 제목(열)** 은 `title`·`genre`·`rating` 처럼 항목 이름입니다.

- **행 하나** = 딕셔너리 (`{"title": "인터스텔라", "rating": 8.7, ...}`)
- **표 전체** = 그 딕셔너리들을 순서대로 담은 리스트

이렇게 **리스트 안에 딕셔너리를 담는 구조**는 표 데이터를 표현하는 가장 기본적인 방법입니다. 아래 `movies` 는 영화 10편의 정보입니다.

````python
movies = [
    {"title": "인터스텔라",   "genre": "SF",     "year": 2014, "rating": 8.7, "runtime": 169},
    {"title": "기생충",       "genre": "드라마", "year": 2019, "rating": 8.9, "runtime": 132},
    {"title": "어벤져스",     "genre": "액션",   "year": 2012, "rating": 8.0, "runtime": 143},
    {"title": "토이 스토리",  "genre": "애니",   "year": 1995, "rating": 8.3, "runtime":  81},
    {"title": "다크 나이트",  "genre": "액션",   "year": 2008, "rating": 9.0, "runtime": 152},
    {"title": "라라랜드",     "genre": "드라마", "year": 2016, "rating": 8.0, "runtime": 128},
    {"title": "마션",         "genre": "SF",     "year": 2015, "rating": 8.0, "runtime": 144},
    {"title": "코코",         "genre": "애니",   "year": 2017, "rating": 8.4, "runtime": 105},
    {"title": "조커",         "genre": "드라마", "year": 2019, "rating": 8.4, "runtime": 122},
    {"title": "인셉션",       "genre": "SF",     "year": 2010, "rating": 8.8, "runtime": 148},
]

print("영화 편수:", len(movies))   # 10
print(movies[0]["title"])         # 행 -> 열 순으로 접근: 인터스텔라
print(movies[1]["rating"])        # 8.9
````

위 출력처럼, **먼저 몇 번째 행인지**(`movies[0]`) 고른 뒤 **어떤 열인지**(`["title"]`) 지정해 값 하나를 꺼냅니다. 표에서 "첫 번째 행, 제목 칸"을 짚는 것과 같습니다.

## 2. 전체 순회와 특정 열(열 이름=키) 추출

표의 모든 행을 차례로 살펴볼 때는 `for` 반복문을 씁니다. 반복할 때마다 변수(`m`)에 행 하나(딕셔너리)가 담깁니다.

````python
# 모든 영화의 제목과 평점을 한 줄씩 출력
for m in movies:
    print(m["title"], "-", m["rating"])

# 인터스텔라 - 8.7
# 기생충 - 8.9
# 어벤져스 - 8.0
# ... (10편 모두 출력됩니다)
````

### 2-1) 특정 열만 모아 새 리스트로

"제목 열만" 또는 "평점 열만" 따로 모으고 싶을 때가 많습니다. 빈 리스트를 만들고 반복하며 `append` 해도 되고, 한 줄로 쓰는 **리스트 컴프리헨션**을 써도 됩니다(2차시에서 다룬 표현입니다).

````python
# 방법 1) 반복문으로 하나씩 모으기
titles = []
for m in movies:
    titles.append(m["title"])
print(titles)
# ["인터스텔라", "기생충", "어벤져스", "토이 스토리", "다크 나이트", "라라랜드", "마션", "코코", "조커", "인셉션"]

# 방법 2) 리스트 컴프리헨션 - 같은 일을 한 줄로
ratings = [m["rating"] for m in movies]
print(ratings)
# [8.7, 8.9, 8.0, 8.3, 9.0, 8.0, 8.0, 8.4, 8.4, 8.8]
````

## 3. 합계·평균·최댓값·최솟값

평점 열을 리스트로 모아 두면, 1차시·이전 노트북에서 본 `sum`·`len`·`max`·`min` 을 그대로 쓸 수 있습니다. **평균은 합계 ÷ 개수** 로 직접 구합니다.

````python
ratings = [m["rating"] for m in movies]

total = sum(ratings)
count = len(ratings)
average = total / count          # 평균 = 합계 / 개수

print("합계:", round(total, 2))      # 84.5  (round: 소수 둘째 자리 반올림)
print("편수:", count)                # 10
print("평균:", round(average, 2))    # 8.45
print("최고 평점:", max(ratings))    # 9.0
print("최저 평점:", min(ratings))    # 8.0
````

### 3-1) 최고 평점인 영화의 '제목'까지 찾기

`max(ratings)` 는 가장 큰 **점수(9.0)** 만 알려 줄 뿐, 그게 어떤 영화인지는 말해 주지 않습니다. 제목까지 알려면 모든 행을 돌며 가장 높은 평점을 가진 행을 직접 찾아야 합니다.

````python
best = movies[0]                 # 일단 첫 영화를 후보로 둠
for m in movies:
    if m["rating"] > best["rating"]:   # 더 높은 평점을 만나면
        best = m                       # 후보 교체

print("최고 평점 영화:", best["title"], best["rating"])
# 최고 평점 영화: 다크 나이트 9.0
````

## 4. 조건으로 거르기(필터링)

"평점 8.5 이상인 영화만" 또는 "SF 장르만" 처럼 조건에 맞는 행만 골라내는 작업입니다. 빈 리스트를 만들고, 반복하며 **조건이 참인 행만** `append` 합니다.

````python
# 평점 8.5 이상인 영화만 골라내기
high = []
for m in movies:
    if m["rating"] >= 8.5:
        high.append(m)

for m in high:
    print(m["title"], m["rating"])
# 인터스텔라 8.7
# 기생충 8.9
# 다크 나이트 9.0
# 인셉션 8.8
````

### 4-1) 리스트 컴프리헨션 + `if`

필터링도 컴프리헨션 한 줄로 쓸 수 있습니다. `for` 뒤에 `if` 조건을 붙이면 조건이 참인 행만 모입니다.

````python
# SF 장르 영화의 '제목'만 한 줄로 모으기
sf_titles = [m["title"] for m in movies if m["genre"] == "SF"]
print(sf_titles)
# ["인터스텔라", "마션", "인셉션"]

# 2015년 이후(>=) 개봉작 편수 세기
recent = [m for m in movies if m["year"] >= 2015]
print("2015년 이후 개봉작:", len(recent), "편")   # 2015년 이후 개봉작: 5 편
````

## 5. 정렬 — `sorted` 와 람다(lambda)

리스트의 `.sort()` 는 숫자·문자 리스트를 정렬할 때 좋지만, 딕셔너리가 든 표는 **"어떤 열을 기준으로" 정렬할지** 알려 줘야 합니다. 이때 `sorted(데이터, key=기준)` 를 쓰고, `key` 에는 **"행 하나를 받아 기준 값을 돌려주는 짧은 함수"** 를 줍니다.

그 짧은 함수를 매번 `def` 로 만들기는 번거로우므로, **람다(lambda)** 라는 한 줄 함수를 씁니다.

```
lambda m: m["rating"]      # m 을 받아 m["rating"] 을 돌려주는 함수
```

이는 아래 `def` 와 같은 뜻입니다.

```
def get_rating(m):
    return m["rating"]
```

````python
# 평점이 높은 순으로 정렬 (reverse=True 면 내림차순)
by_rating = sorted(movies, key=lambda m: m["rating"], reverse=True)

for m in by_rating:
    print(m["rating"], m["title"])
# 9.0 다크 나이트
# 8.9 기생충
# 8.8 인셉션
# ... (평점 높은 순으로 10편)
````

### 5-1) 다른 기준으로 정렬

`key` 의 람다만 바꾸면 어떤 열로든 정렬할 수 있습니다. `sorted` 는 **원본 `movies` 를 바꾸지 않고** 정렬된 새 리스트를 돌려줍니다.

````python
# 개봉연도가 오래된 순(오름차순)
by_year = sorted(movies, key=lambda m: m["year"])
print([m["year"] for m in by_year])
# [1995, 2008, 2010, 2012, 2014, 2015, 2016, 2017, 2019, 2019]

# 상영시간이 짧은 순 - 상위 3편만 슬라이싱
shortest3 = sorted(movies, key=lambda m: m["runtime"])[:3]
print([m["title"] for m in shortest3])
# ["토이 스토리", "코코", "조커"]
````

## 6. 그룹별로 세기 — 딕셔너리 집계

"장르별로 영화가 몇 편씩 있는지" 같은 집계는 표 데이터에서 매우 자주 나옵니다. **결과를 담을 딕셔너리**를 하나 만들고, 모든 행을 돌며 **그 장르의 개수를 1씩 늘립니다.**

문제는 "아직 한 번도 안 나온 장르"입니다. 처음 보는 키는 딕셔너리에 없어서 바로 `+1` 하면 오류가 납니다. 이를 깔끔하게 처리하는 두 가지 방법을 봅니다.

### 6-1) `get(키, 기본값)` 으로 세기

`counts.get(genre, 0)` 은 "그 키가 있으면 값을, 없으면 0을" 돌려줍니다. 처음 보는 장르도 0에서 시작해 안전하게 1을 더할 수 있습니다.

````python
counts = {}
for m in movies:
    genre = m["genre"]
    counts[genre] = counts.get(genre, 0) + 1   # 없으면 0에서 시작, +1

print(counts)   # {"SF": 3, "드라마": 3, "액션": 2, "애니": 2}
````

### 6-2) `setdefault(키, 기본값)` 으로 묶기

개수만이 아니라 **그룹에 속한 항목 자체를 모으고** 싶을 때가 있습니다. `setdefault(genre, [])` 는 "그 키가 없으면 빈 리스트를 먼저 넣어 준다"는 뜻이라, 곧바로 `append` 할 수 있습니다.

````python
groups = {}                      # 장르 -> 제목 리스트
for m in movies:
    genre = m["genre"]
    groups.setdefault(genre, [])        # 없으면 빈 리스트 준비
    groups[genre].append(m["title"])

for genre, titles in groups.items():
    print(genre, ":", titles)
````

### 6-3) 그룹별 평균 평점

6-2의 묶기와 합계·평균을 합치면 "장르별 평균 평점"도 구할 수 있습니다. 장르별로 평점을 리스트에 모은 뒤, 각 리스트의 평균을 계산합니다.

````python
# 1) 장르별로 평점을 모은다
ratings_by_genre = {}
for m in movies:
    ratings_by_genre.setdefault(m["genre"], [])
    ratings_by_genre[m["genre"]].append(m["rating"])

# 2) 각 장르의 평균을 구한다
for genre, rs in ratings_by_genre.items():
    avg = sum(rs) / len(rs)
    print(genre, "평균 평점:", round(avg, 2))
# SF 평균 평점: 8.5
# 드라마 평균 평점: 8.43
# 액션 평균 평점: 8.5
# 애니 평균 평점: 8.35
````

## 7. 미리 보기 — 4차시에서는 pandas로 한두 줄

방금 우리는 표 데이터를 다루려고 **반복문·조건·딕셔너리 집계**를 여러 줄에 걸쳐 작성했습니다. 합계 하나, 정렬 하나마다 코드가 늘어났습니다.

4차시에서 배울 `pandas` 는 이런 표 데이터를 전문으로 다루는 도구입니다. 같은 작업이 얼마나 짧아지는지 아래에서 비교해 보세요. **지금은 개념만 눈으로 확인**하고, 실제 실행은 4차시에서 합니다(그래서 아래 코드는 `#` 으로 주석 처리해 두었습니다).

### 우리가 한 작업 vs pandas

| 작업 | 오늘(순수 파이썬) | 4차시(pandas) |
|---|---|---|
| 평점 평균 | 평점을 리스트로 모아 `sum/len` | `df["rating"].mean()` |
| 평점 8.5 이상 필터 | 빈 리스트 + 반복 + `if` + `append` | `df[df["rating"] >= 8.5]` |
| 평점 높은 순 정렬 | `sorted(..., key=lambda ...)` | `df.sort_values("rating")` |
| 장르별 개수 | 딕셔너리 + `get` 으로 누적 | `df["genre"].value_counts()` |
| 장르별 평균 평점 | 묶고 → 각 평균 계산 | `df.groupby("genre")["rating"].mean()` |

````python
# 아래는 4차시용 미리 보기입니다. (지금은 실행하지 않습니다 - 모두 주석 처리)
#
# import pandas as pd
#
# df = pd.DataFrame(movies)          # 리스트 of 딕셔너리를 그대로 표로
# print(df)                          # 표 전체를 깔끔하게 출력
#
# df["rating"].mean()                # 평점 평균  (오늘은 여러 줄이었던 작업)
# df[df["rating"] >= 8.5]            # 평점 8.5 이상 필터
# df.sort_values("rating", ascending=False)   # 평점 높은 순 정렬
# df["genre"].value_counts()         # 장르별 개수
# df.groupby("genre")["rating"].mean()         # 장르별 평균 평점

print("pandas 미리 보기 - 자세한 실행은 4차시에서 다룹니다.")
````

## 정리
- 표 데이터는 **리스트 안에 딕셔너리**로 표현합니다. 행 하나가 딕셔너리, 표 전체가 리스트입니다.
- 합계·평균·최댓값은 **열을 리스트로 모아** `sum`·`len`·`max`·`min` 으로 구합니다.
- **필터링**은 반복 + `if`(또는 컴프리헨션 + `if`), **정렬**은 `sorted(데이터, key=lambda ...)` 로 합니다.
- **그룹 집계**는 빈 딕셔너리에 `get`·`setdefault` 패턴으로 누적합니다.
- 이 모든 작업을 4차시에서는 **`pandas`** 로 한두 줄에 처리합니다. 오늘의 수고가 그 편리함을 이해하는 바탕이 됩니다.
