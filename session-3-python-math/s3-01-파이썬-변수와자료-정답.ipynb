## 1. 변수 — 값에 이름 붙이기
변수는 값에 이름을 붙여 저장해 두는 공간입니다. `=` 기호는 "오른쪽 값을 왼쪽 이름에 저장하라"는 의미로, 수학의 등호(같다)와는 다릅니다. 저장한 값은 이름으로 언제든 다시 꺼내 쓰거나 새로 바꿀 수 있습니다.

```python
course = "인공지능 입문"   # 문자열을 course 에 저장
year = 2026               # 정수를 year 에 저장

print(course)             # 인공지능 입문
print(year)               # 2026

year = year + 1           # 기존 값에 1을 더해 다시 저장
print(year)               # 2027
```

### 변수 이름 규칙
- 영문자·숫자·밑줄(`_`)로 짓고, **숫자로 시작할 수 없습니다.**
- 대소문자를 구분합니다 (`score` 와 `Score` 는 다른 변수).
- 의미가 드러나는 이름을 권장합니다 (`a` 보다 `student_count`).

```python
student_count = 30
average_score = 87.5
print(student_count, average_score)   # 30 87.5
```

## 2. 자료형 — 값의 종류
파이썬의 기본 자료형은 다음과 같습니다.

| 자료형 | 설명 | 예 |
|---|---|---|
| `int` | 정수 | `1`, `-5`, `100` |
| `float` | 실수 | `3.14`, `-0.5` |
| `str` | 문자열 | `"AI"`, `'한경대'` |
| `bool` | 불리언(참/거짓) | `True`, `False` |

`type()` 으로 값의 자료형을 확인합니다.

```python
a = 100          # int
b = 3.14         # float
c = "Hello"      # str
d = True         # bool

print(type(a))   # <class 'int'>
print(type(b))   # <class 'float'>
print(type(c))   # <class 'str'>
print(type(d))   # <class 'bool'>
```

## 3. 형 변환 — 자료형 바꾸기
자료형은 서로 바꿀 수 있습니다. `int()` 정수로, `float()` 실수로, `str()` 문자열로 변환합니다. 특히 화면 입력값은 항상 문자열이므로 계산하려면 숫자로 바꿔야 합니다.

```python
num_str = "25"           # 문자열 "25"
print(num_str + "년")     # 25년   (문자열 이어붙이기)

num = int(num_str)       # 정수 25 로 변환
print(num + 5)           # 30     (숫자 계산)

print(str(num) + "살")    # 25살   (다시 문자열로)
```

### 참고: 사용자 입력 받기

```python
height_str = input("키를 입력하세요(cm): ")
height = float(height_str)
print(f"입력한 키는 {height}cm 입니다.")
```

## 4. 연산자
### 4-1) 산술 연산자

```python
a = 17
b = 5
print("덧셈:", a + b)      # 22
print("뺄셈:", a - b)      # 12
print("곱셈:", a * b)      # 85
print("나눗셈:", a / b)    # 3.4
print("몫:", a // b)       # 3   (소수점 버림)
print("나머지:", a % b)    # 2
print("거듭제곱:", a ** b) # 1419857
```

### 4-2) 비교 연산자
비교의 결과는 항상 `True`/`False` 입니다. `==` 는 '같다', `=` 는 '저장'으로 서로 다릅니다.

```python
x = 10
y = 7
print(x > y)    # True
print(x == y)   # False
print(x != y)   # True   (!= 는 '다르다')
print(x >= 10)  # True
```

### 4-3) 논리 연산자
`and`(둘 다 참), `or`(하나라도 참), `not`(반대).

```python
age = 22
has_id = True
print(age >= 19 and has_id)   # True   (성인이고 신분증도 있음)
print(age < 19 or has_id)     # True
print(not has_id)             # False
```

## 5. 문자열 다루기
### 5-1) 이어붙이기와 f-string
`+` 로 문자열을 잇고, `*` 로 반복합니다. 값을 글자 사이에 끼워 넣을 땐 f-string(따옴표 앞 `f`, `{ }` 안에 값)이 편리합니다.

```python
first = "한경"
last = "국립대"
print(first + last)     # 한경국립대
print(first * 3)        # 한경한경한경

name = "지민"
score = 92
print(f"{name}님의 점수는 {score}점입니다.")
```

### 5-2) 인덱싱과 슬라이싱
문자열의 각 글자에는 0번부터 번호가 붙습니다. `[시작:끝]` 으로 일부를 잘라냅니다(끝 번호는 포함하지 않음). 음수 인덱스는 뒤에서부터 셉니다.

```python
text = "Artificial"
print(text[0])     # A     (0번째)
print(text[-1])    # l     (마지막)
print(text[0:4])   # Arti  (0~3번)
print(text[:4])    # Arti
print(text[4:])    # ficial
print(len(text))   # 10    (글자 수)
```

### 5-3) 문자열 메서드
메서드는 값에 `.` 을 붙여 호출하는 기능입니다. 문자열에는 자주 쓰는 메서드가 많습니다.

```python
s = "  Hello, AI World  "
print(s.strip())              # 양옆 공백 제거
print(s.upper())              # 대문자
print(s.lower())              # 소문자
print(s.replace("AI", "ML"))  # 일부 바꾸기
print("a,b,c".split(","))     # ['a', 'b', 'c']  쪼개기
```

## 6. 리스트 — 여러 값을 순서대로
대괄호 `[ ]` 로 만들며, 0번부터 번호(인덱스)로 접근합니다. 문자열처럼 슬라이싱도 됩니다.

```python
scores = [88, 92, 79, 95, 60]
print(scores[0])      # 88
print(scores[-1])     # 60
print(scores[1:3])    # [92, 79]   (슬라이싱)
print(len(scores))    # 5
print(sum(scores))    # 414
print(max(scores), min(scores))   # 95 60
```

### 6-1) 리스트 메서드

```python
fruits = ["사과", "바나나", "체리"]
fruits.append("오렌지")      # 맨 뒤에 추가
fruits.insert(0, "딸기")     # 0번 위치에 삽입
fruits.remove("바나나")      # 값으로 삭제
print(fruits)               # ['딸기', '사과', '체리', '오렌지']

fruits.sort()               # 가나다순 정렬
print(fruits)
```

### 연습문제 4
- `nums = [5, 3, 9, 1, 7]` 을 만드세요.
- 정렬한 결과, 합계(`sum`), 평균(합계/개수)을 각각 출력하세요.

```python
nums = [5, 3, 9, 1, 7]
nums.sort()
print(nums)                   # [1, 3, 5, 7, 9]
print(sum(nums))              # 25
print(sum(nums) / len(nums))  # 5.0
```

## 7. 튜플 — 바뀌지 않는 묶음
소괄호 `( )` 로 만들며, 한 번 만들면 **수정할 수 없습니다**. 좌표처럼 고정된 값에 적합합니다.

```python
point = (37.5, 127.0)   # (위도, 경도)
print(point[0])         # 37.5

# point[0] = 0          # 실행하면 오류! 튜플은 수정 불가

lat, lon = point        # 각 값을 변수로 풀기(언패킹)
print(lat, lon)         # 37.5 127.0
```

## 8. 딕셔너리 — 이름표(키)로 값 관리
중괄호 `{ }` 안에 `키: 값` 형태로 저장하고, 키로 값을 꺼냅니다. 리스트가 번호로 접근한다면 딕셔너리는 이름으로 접근합니다.

```python
student = {"name": "지민", "age": 22, "major": "AI"}
print(student["name"])     # 지민
student["grade"] = 3        # 새 항목 추가
student["age"] = 23         # 기존 값 수정
print(student)
```

### 8-1) 키·값 다루기

```python
print(student.keys())            # 키 목록
print(student.values())          # 값 목록
print(student.get("major"))      # AI
print(student.get("phone", "정보 없음"))   # 없는 키는 기본값 반환
```

### 연습문제 6
아래 `students` 는 학생 3명의 정보입니다.
- 각 학생의 이름과 점수를 한 줄씩 출력하고,
- 세 점수의 평균을 구해 출력하세요.

(아직 반복문을 배우지 않았으니 인덱스로 하나씩 꺼내 써도 됩니다. 반복문은 다음 노트북에서 다룹니다.)

```python
students = [
    {"name": "지민", "score": 88},
    {"name": "현우", "score": 92},
    {"name": "수아", "score": 79},
]
# 힌트: students[0]["name"], students[0]["score"]
# 아래에 작성하세요

print(f"{students[0]['name']}: {students[0]['score']}점")
print(f"{students[1]['name']}: {students[1]['score']}점")
print(f"{students[2]['name']}: {students[2]['score']}점")
avg = (students[0]["score"] + students[1]["score"] + students[2]["score"]) / 3
print("평균:", avg)
```

- **문자열**은 인덱싱·슬라이싱·메서드로 자유롭게 가공할 수 있습니다.
