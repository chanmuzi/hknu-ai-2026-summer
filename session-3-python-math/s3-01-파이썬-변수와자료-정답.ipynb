## 1. 변수 — 값에 이름 붙이기
변수는 값에 이름을 붙여 저장해 두는 공간입니다. `=` 기호는 "오른쪽 값을 왼쪽 이름에 저장하라"는 의미로, 수학의 등호(같다)와는 다릅니다. 저장한 값은 이름으로 언제든 다시 꺼내 쓰거나 새로 바꿀 수 있습니다.

```python
course = "인공지능 입문"   # 문자열 저장
year = 2026               # 정수 저장
print(course)
print(year)

year = year + 1           # 기존 값에 1을 더해 다시 저장
print(year)               # 2027
```

### 변수 이름 규칙
- 영문자·숫자·밑줄(`_`)로 짓고, **숫자로 시작할 수 없습니다.**
- 대소문자를 구분합니다 (`score` 와 `Score` 는 다른 변수).
- 의미가 드러나는 이름을 권장합니다 (`a` 보다 `student_count`).

```python
student_count = 30
average_score = 87.5
print(student_count, average_score)   # 30 87.5
```

## 2. 자료형 — 값의 종류
파이썬의 기본 자료형은 다음과 같습니다.

| 자료형 | 설명 | 예 |
|---|---|---|
| `int` | 정수 | `1`, `-5`, `100` |
| `float` | 실수 | `3.14`, `-0.5` |
| `str` | 문자열 | `"AI"`, `'한경대'` |
| `bool` | 불리언(참/거짓) | `True`, `False` |

`type()` 으로 값의 자료형을 확인합니다.

```python
a = 100          # int
b = 3.14         # float
c = "Hello"      # str
d = True         # bool

# type() 으로 자료형 확인
print(type(a), type(b), type(c), type(d))
```

## 3. 형 변환 — 자료형 바꾸기
자료형은 서로 바꿀 수 있습니다. `int()` 정수로, `float()` 실수로, `str()` 문자열로 변환합니다. 특히 화면 입력값은 항상 문자열이므로 계산하려면 숫자로 바꿔야 합니다.

```python
num_str = "25"           # 문자열
print(num_str + "년")     # 문자열 이어붙이기 → 25년

num = int(num_str)       # 정수로 변환
print(num + 5)           # 숫자 계산 → 30

print(str(num) + "살")    # 다시 문자열로 → 25살
```

## 4. 연산자
### 4-1) 산술 연산자

```python
a = 17
b = 5
print("덧셈:", a + b)
print("뺄셈:", a - b)
print("곱셈:", a * b)
print("나눗셈:", a / b)
print("몫:", a // b)       # 소수점 버림
print("나머지:", a % b)
print("거듭제곱:", a ** b)
```

### 4-2) 비교 연산자
비교의 결과는 항상 `True`/`False` 입니다. `==` 는 '같다', `=` 는 '저장'으로 서로 다릅니다.

```python
x = 10
y = 7
print(x > y)
print(x == y)   # == 는 '같다'
print(x != y)   # != 는 '다르다'
print(x >= 10)
```

### 4-3) 논리 연산자
`and`(둘 다 참), `or`(하나라도 참), `not`(반대).

```python
age = 22
has_id = True
print(age >= 19 and has_id)   # 둘 다 참
print(age < 19 or has_id)     # 하나라도 참
print(not has_id)             # 반대
```

## 5. 문자열 다루기
### 5-1) 이어붙이기와 f-string
`+` 로 문자열을 잇고, `*` 로 반복합니다. 값을 글자 사이에 끼워 넣을 땐 f-string(따옴표 앞 `f`, `{ }` 안에 값)이 편리합니다.

```python
first = "한경"
last = "국립대"
print(first + last)     # 이어붙이기 → 한경국립대
print(first * 3)        # 반복 → 한경한경한경

name = "지민"
score = 92
print(f"{name}님의 점수는 {score}점입니다.")   # f-string
```

### 5-2) 인덱싱과 슬라이싱
문자열의 각 글자에는 0번부터 번호가 붙습니다. `[시작:끝]` 으로 일부를 잘라냅니다(끝 번호는 포함하지 않음). 음수 인덱스는 뒤에서부터 셉니다.

```python
text = "Artificial"
print(text[0])     # 0번째 → A
print(text[-1])    # 마지막 → l
print(text[0:4])   # 0~3번 → Arti
print(text[:4])    # Arti
print(text[4:])    # ficial
print(len(text))   # 글자 수 → 10
```

### 5-3) 문자열 메서드
메서드는 값에 `.` 을 붙여 호출하는 기능입니다. 문자열에는 자주 쓰는 메서드가 많습니다.

```python
s = "  Hello, AI World  "
print(s.strip())              # 양옆 공백 제거
print(s.upper())              # 대문자
print(s.lower())              # 소문자
print(s.replace("AI", "ML"))  # 일부 바꾸기
print("a,b,c".split(","))     # 쪼개기 → ['a','b','c']
```

## 6. 리스트 — 여러 값을 순서대로
대괄호 `[ ]` 로 만들며, 0번부터 번호(인덱스)로 접근합니다. 문자열처럼 슬라이싱도 됩니다.

```python
scores = [88, 92, 79, 95, 60]
print(scores[0])      # 첫 항목 → 88
print(scores[-1])     # 마지막 → 60
print(scores[1:3])    # 슬라이싱 → [92, 79]
print(len(scores))    # 개수 → 5
print(sum(scores))    # 합 → 414
print(max(scores), min(scores))   # 95 60
```

### 6-1) 리스트 메서드

```python
fruits = ["사과", "바나나", "체리"]
fruits.append("오렌지")      # 맨 뒤에 추가
fruits.insert(0, "딸기")     # 0번 위치에 삽입
fruits.remove("바나나")      # 값으로 삭제
print(fruits)

fruits.sort()               # 가나다순 정렬
print(fruits)
```

<b>문제</b> &nbsp; <code>nums = [5, 3, 9, 1, 7]</code> 의 정렬 결과·합계(<code>sum</code>)·평균을 출력하세요.<br>

## 7. 튜플 — 바뀌지 않는 묶음
소괄호 `( )` 로 만들며, 한 번 만들면 **수정할 수 없습니다**. 좌표처럼 고정된 값에 적합합니다.

```python
point = (37.5, 127.0)   # (위도, 경도)
print(point[0])         # 37.5

# point[0] = 0          # 튜플은 수정 불가 → 오류

lat, lon = point        # 언패킹
print(lat, lon)
```

## 8. 딕셔너리 — 이름표(키)로 값 관리
중괄호 `{ }` 안에 `키: 값` 형태로 저장하고, 키로 값을 꺼냅니다. 리스트가 번호로 접근한다면 딕셔너리는 이름으로 접근합니다.

```python
student = {"name": "지민", "age": 22, "major": "AI"}
print(student["name"])     # 키로 값 꺼내기 → 지민
student["grade"] = 3        # 새 항목 추가
student["age"] = 23         # 기존 값 수정
print(student)
```

### 8-1) 키·값 다루기

```python
print(student.keys())                     # 키 목록
print(student.values())                   # 값 목록
print(student.get("major"))               # AI
print(student.get("phone", "정보 없음"))   # 없는 키는 기본값 반환
```

## 9. 종합 연습
지금까지 배운 변수·자료형·리스트·딕셔너리를 함께 사용해 봅니다.

<div style="border:1px solid #E7DCC6;border-left:6px solid #C15F3C;border-radius:10px;background:#FBF4EA;padding:16px 20px;margin:12px 0;">
<div style="font-size:16px;font-weight:700;color:#A8472A;margin-bottom:8px;">✏️ 종합문제 학생 점수 다루기 <span style="font-size:12px;color:#A8472A;background:#F0DCCF;padding:1px 8px;border-radius:10px;margin-left:6px;">종합</span></div>
<div style="color:#3D3A33;line-height:1.8;font-size:14.5px;">
<b>문제</b> &nbsp; <code>students</code>(학생 3명)의 이름·점수를 한 줄씩 출력하고, 세 점수의 평균을 구해 출력하세요.<br>
<b>힌트</b> &nbsp; 인덱스로 꺼내기 <code>students[0]["name"]</code> · 반복문은 다음 노트북에서 배웁니다.<br>
<b>예상</b> &nbsp; 이름: 점수 3줄 + 평균 약 86.3
</div>
</div>

```python
students = [
    {"name": "지민", "score": 88},
    {"name": "현우", "score": 92},
    {"name": "수아", "score": 79},
]
# ✅ 종합문제 정답
print(f"{students[0]['name']}: {students[0]['score']}점")
print(f"{students[1]['name']}: {students[1]['score']}점")
print(f"{students[2]['name']}: {students[2]['score']}점")
avg = (students[0]["score"] + students[1]["score"] + students[2]["score"]) / 3
print("평균:", avg)
```

## 정리
- **변수**는 `=` 로 값에 이름을 붙이는 것이며, 기본 자료형은 `int`·`float`·`str`·`bool` 입니다.
- **문자열**은 인덱싱·슬라이싱·메서드로 자유롭게 가공할 수 있습니다.
- **리스트**(`[ ]`, 순서) · **튜플**(`( )`, 고정) · **딕셔너리**(`{ }`, 키-값)로 여러 값을 묶어 관리합니다.
