### Numpy 추가 예제

```python
import numpy as np
```

```python
# 1. array — 리스트를 배열로
arr = np.array([1, 2, 3, 4, 5])
print('arr:\n', arr)
```

```python
# 2. arange — 연속된 수열
rng = np.arange(0, 10, 2)
print('rng:\n', rng)
```

```python
# 3. linspace — 등간격 숫자
lin = np.linspace(0, 1, 4)
print('lin:\n', lin)
```

```python
# 4. zeros — 0으로 채운 배열
z = np.zeros((2, 3))
print('zeros:\n', z)
```

```python
# 5. ones — 1로 채운 배열
o = np.ones((3, 2))
print('ones:\n', o)
```

```python
# 6. eye — 단위행렬
I = np.eye(4)
print('eye:\n', I)
```

```python
# 7. random.seed — 난수 고정(재현용)
np.random.seed(42)
```

```python
# 8. random.rand — 균등분포 난수
r1 = np.random.rand(2, 2)
print('rand:\n', r1)
```

```python
# 9. random.randn — 정규분포 난수
r2 = np.random.randn(3, 3)
print('randn:\n', r2)
```

```python
# 10. reshape — 형태 변경
arr = np.array([[1, 2, 3, 4, 5], [7,9,11,13,15]])
reshaped = arr.reshape((5, 2))
print('reshaped:\n', reshaped)
```

```python
# 11. flatten — 1차원으로 펼치기
flat = r2.flatten()
print('flat:\n', flat)
```

```python
# 12. concatenate — 이어 붙이기
a = np.array([1, 2]); b = np.array([3, 4])
cat = np.concatenate([a, b])
print('concatenate:\n', cat)
```

```python
# 13. vstack — 위아래로 쌓기
v = np.vstack([a, b])
print('vstack:\n', v)
```

```python
# 14. hstack — 좌우로 붙이기
h = np.hstack([a.reshape(2,1), b.reshape(2,1)])
print('hstack:\n', h)
```

```python
# 15. unique — 고유값
uniq = np.unique([1,2,2,3,3,3])
print('unique:\n', uniq)
```

```python
# 16. argmax — 최대값 위치
idx_max = np.argmax([1,5,2,4])
print('argmax:\n', idx_max)
```

```python
# 17. where — 조건에 맞는 위치
w = np.where(arr % 2 == 0)
print('where (even indices):\n', w)
```

```python
# 18. clip — 값 범위 제한
cl = np.clip(rng, 1, 7)
print('clip:\n', cl)
```

```python
# 19. broadcast_to — 브로드캐스팅
bcd = np.broadcast_to([1,2,3], (3,3))
print('broadcast_to:\n', bcd)
```

```python
# 20. transpose — 축 바꾸기
t = np.transpose(r2, (1,0))
print('transpose:\n', t)
```

### Pandas 추가 예제

```python
import pandas as pd

# 샘플 데이터프레임 생성
data = {
    '이름': ['홍길동', '이영희', '김철수', '박민수'],
    '나이': [25, 30, None, 22],
    '성별': ['M', 'F', 'M', 'M'],
    '점수': [88, 92, 79, None]
}
df = pd.DataFrame(data)
print(df)
```

```python
# 1. Series — 1차원 데이터
s = pd.Series([1, 2, 3])
print('Series:\n', s)
```

```python
# 2. read_csv / read_excel — 파일에서 읽기 (예시, 실제 파일 없으면 실행 안 함)
# df_csv = pd.read_csv('파일.csv')
# df_excel = pd.read_excel('파일.xlsx', sheet_name='1번시트')
```

```python
# 3. head — 위에서 몇 줄
print('\nhead():\n', df.head())
```

```python
# 4. tail — 아래에서 몇 줄
print('\ntail():\n', df.tail())
```

```python
# 5. info — 열·타입·결측 요약
print('\ninfo():')
df.info()
```

```python
# 6. describe — 숫자열 통계 요약
print('\ndescribe():\n', df.describe())
```

```python
# 7. columns — 열 이름
print('\ncolumns:', df.columns)
```

```python
# 8. index — 행 인덱스
print('\nindex:', df.index)
```

```python
# 9. shape — (행, 열) 크기
print('\nshape:', df.shape)
```

```python
# 10. dtypes — 열별 타입
print('\ndtypes:\n', df.dtypes)
```

```python
# 11. sort_values — 값 기준 정렬
print('\nsort_values(score):\n', df.sort_values(by='점수'))
```

```python
# 12. filter — 열 골라내기
print('\nfilter(columns):\n', df.filter(items=['이름','점수']))
```

```python
# 13. groupby — 그룹으로 묶기
grp = df.groupby('성별')
print('\ngroupby:', grp)
```

```python
# 14. agg — 그룹별 집계
print('\nagg mean:\n', grp.agg({'나이':'mean','점수':'mean'}))
```

```python
# 15. merge — 키로 합치기
df2 = pd.DataFrame({'이름':['홍길동','이영희'],'도시':['서울','부산']})
print('\nmerge:\n', pd.merge(df, df2, on='이름', how='left'))
```

```python
# 16. concat — 이어 붙이기
print('\nconcat axis=1:\n', pd.concat([df, df.filter(items=['이름','점수'])], axis=1))
```

```python
# 17. apply — 열에 함수 적용
df['점수2'] = df['점수'].apply(lambda x: x**2 if pd.notnull(x) else x)
print('\napply square:\n', df)
```

```python
# 18. isnull — 결측 여부
print('\nisnull:\n', df.isnull())
```

```python
# 19. dropna — 결측 행 제거
print('\ndropna:\n', df.dropna(subset=['나이','점수']))
```

```python
# 20. fillna — 결측 채우기
print('\nfillna:\n', df.fillna({'나이':df['나이'].mean(),'점수':df['점수'].mean()}))
```

```python
# 21. astype — 타입 변경
print('\nastype:\n', df['나이'].astype('float'))
```

```python
# 22. rename — 열 이름 변경
print('\nrename columns:\n', df.rename(columns={'점수':'score'}))
```

```python
# 23. drop — 행·열 제거
print('\ndrop column:\n', df.drop('점수2', axis=1))
```

```python
# 24. sample — 무작위 표본
print('\nsample 2 rows:\n', df.sample(2))
```

```python
# 25. pivot_table — 피벗 표
print('\npivot_table:\n', df.pivot_table(index='성별', values='점수', aggfunc='mean'))
```

```python
# 26. get_dummies — 원-핫 인코딩
print('\nget_dummies:\n', pd.get_dummies(df['성별']))
```

```python
# 27. to_csv: CSV 저장
df.to_csv('df_example.csv', index=False)
print("\nSaved DataFrame to 'df_example.csv'")
```

```python
# 29. drop_duplicates: 중복 제거
print('\ndrop_duplicates:\n', df.drop_duplicates())
```

```python
# 30. sort_index: 인덱스 기준 정렬
print('\nsort_index:\n', df.sort_index())
```
