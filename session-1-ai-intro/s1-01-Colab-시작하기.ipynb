## 1. Colab이란?
Google Colab은 **브라우저에서 파이썬을 실행**하는 무료 도구입니다. 내 컴퓨터에 설치할 것이 없고, 구글 계정만 있으면 됩니다. 코드는 구글의 컴퓨터에서 실행되고 결과만 화면에 표시됩니다.

## 2. 셀 실행하기
Colab은 **셀**(칸) 단위로 실행합니다. 셀을 클릭하고 `Shift + Enter` 또는 셀 왼쪽 **▶ 버튼**을 누르면 실행됩니다. 아래 셀을 실행해 보세요.

```python
print("Colab 실행 성공!")
```

실행되면 셀 아래에 결과가 나타나고, 셀 왼쪽에는 실행 순서 번호(`[1]`)가 표시됩니다.

## 3. 두 가지 셀: 코드 / 텍스트
- **코드 셀**: 코드를 실행하는 칸(아래 회색 칸).
- **텍스트(마크다운) 셀**: 지금 읽고 있는 설명처럼 글을 적는 칸.

상단의 **`+ 코드`**, **`+ 텍스트`** 버튼으로 새 셀을 추가할 수 있습니다.

## 4. 셀 다루기 (추가·삭제·이동)
- **추가**: 셀 사이에 마우스를 올리면 나타나는 `+ 코드` / `+ 텍스트`
- **삭제**: 셀 오른쪽 위 **휴지통** 아이콘
- **이동**: 셀 오른쪽 위 **↑ ↓** 화살표

아래 셀에 직접 계산식을 적고 실행해 보세요.

```python
1 + 1     # 숫자를 바꿔 실행해 보세요
```

## 15. 자주 만나는 오류와 해결
실습 중 빨간 오류 메시지가 떠도 당황하지 마세요. 대부분 아래 네 가지입니다.

| 증상 | 원인 | 해결 |
|---|---|---|
| `NameError: name '...' is not defined` | 그 값을 만든 **앞 셀을 실행하지 않음** | 맨 위부터 순서대로 다시 실행 |
| `ModuleNotFoundError: No module named '...'` | 패키지가 **설치되지 않음** | `!pip install 패키지이름` 먼저 실행 |
| 셀 왼쪽이 `[*]` 인 채 멈춤 | 실행이 **오래 걸리거나 멈춤** | 런타임 → 세션 다시 시작 후 재실행 |
| 한참 기다려도 반응 없음 | 런타임 **연결이 끊김** | 우측 상단 **연결**(또는 다시 연결) 클릭 |

> 오류 메시지의 **맨 아랫줄**이 핵심입니다. 그대로 검색하거나 강사에게 보여주면 빠르게 해결됩니다.
