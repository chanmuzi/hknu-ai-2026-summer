## 1. Colab이란?
Google Colab은 **브라우저에서 파이썬을 실행**하는 무료 도구입니다. 내 컴퓨터에 설치할 것이 없고, 구글 계정만 있으면 됩니다. 코드는 구글의 컴퓨터에서 실행되고 결과만 화면에 표시됩니다.

## 2. 셀 실행하기
Colab은 **셀**(칸) 단위로 실행합니다. 셀을 클릭하고 `Shift + Enter` 또는 셀 왼쪽 **▶ 버튼**을 누르면 실행됩니다. 아래 셀을 실행해 보세요.

```python
print("Colab 실행 성공!")
```

실행되면 셀 아래에 결과가 나타나고, 셀 왼쪽에는 실행 순서 번호(`[1]`)가 표시됩니다.

## 3. 두 가지 셀: 코드 / 텍스트
- **코드 셀**: 코드를 실행하는 칸(아래 회색 칸).
- **텍스트(마크다운) 셀**: 지금 읽고 있는 설명처럼 글을 적는 칸.

상단의 **`+ 코드`**, **`+ 텍스트`** 버튼으로 새 셀을 추가할 수 있습니다.

## 4. 셀 다루기 (추가·삭제·이동)
- **추가**: 셀 사이에 마우스를 올리면 나타나는 `+ 코드` / `+ 텍스트`
- **삭제**: 셀 오른쪽 위 **휴지통** 아이콘
- **이동**: 셀 오른쪽 위 **↑ ↓** 화살표

아래 셀에 직접 계산식을 적고 실행해 보세요.

```python
1 + 1     # 숫자를 바꿔 실행해 보세요
```

## 5. 실행은 위에서 아래 순서로
셀은 **위에서 아래 순서대로** 실행하는 것이 안전합니다. 앞 셀에서 만든 값을 뒤 셀이 사용하기 때문입니다. 아래 두 셀을 순서대로 실행해 보세요.

```python
message = "안녕하세요, 한경국립대"   # 먼저 이 셀을 실행
```

```python
print(message)                     # 그다음 이 셀 (앞 셀을 건너뛰면 오류가 납니다)
```

## 6. 느낌표 `!` 로 시스템 명령 실행하기
코드 셀 맨 앞에 `!` 를 붙이면 파이썬이 아니라 **시스템(리눅스) 명령**이 실행됩니다. 실습에서 패키지를 설치하거나 파일을 확인할 때 자주 씁니다.

```python
!pwd      # print working directory: 현재 작업 폴더 경로
```

```python
!ls       # list: 현재 폴더의 파일 목록
```

```python
!pip install --quiet emoji      # 패키지 설치 예시 (실습에서 자주 보게 됩니다)
```
